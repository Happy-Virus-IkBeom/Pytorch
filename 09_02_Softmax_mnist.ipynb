```python
from __future__ import print_function
from torch import nn, optim, cuda, tensor, eq
from torch.utils import data
from torchvision import datasets, transforms
import torch.nn.functional as F
import time
```

```python
batch_size = 64
device = 'cuda' if cuda.is_available() else 'cpu'
print(f'Training MNIST Model on {device}\n{"=" * 44}')
```

```python
# MNIST Dataset
train_dataset = datasets.MNIST(root='./MNIST_DATASET/',
                               train=True,
                               transform=transforms.ToTensor(),
                               download=True)
```

```python
# view 역할
x = tensor([[1,2,3], 
            [4,5,6]])
x = x.view(-1,6)
print(x)
print(x.size())
```

```python
# torch.eq 역할 -> torch.eq(input, other) : input 과 other의 component 를 비교해서 각각의 component 들이같은지 다른지 boolean 으로 출력해줌.
X = tensor([[1,2,3],
            [4,5,6]])
Y = tensor([[1,6,3],
            [2,8,4]])
print(eq(X,Y))
```

```python
# MNIST Dataset
test_dataset = datasets.MNIST(root='./MNIST_DATASET/',
                               train=False,
                               transform=transforms.ToTensor(),
                               download=True)
```

```python
# Data Loader (Input Pipeline)
train_loader = data.DataLoader(dataset=train_dataset,
                                           batch_size=batch_size,
                                           shuffle=True)

test_loader = data.DataLoader(dataset=test_dataset,
                                          batch_size=batch_size,
                                          shuffle=False)
```

```python
for batch_idx, (data, target) in enumerate(train_loader):
    print(f'batch index: {batch_idx}')
    print(f'data = {data},\n target = {target}')
    print('\n')
```

```python
class Net(nn.Module):
    
    def __init__(self):
        super(Net, self).__init__() #super().__init__ 과 같은거임. 다만, super(파생클래스,self) 를 통해 지금의 클래스를 명확하게 표시하는 역할을 함.
        self.l1 = nn.Linear(784, 520)
        self.l2 = nn.Linear(520, 320)
        self.l3 = nn.Linear(320, 240)
        self.l4 = nn.Linear(240, 120)
        self.l5 = nn.Linear(120, 10)
    
    def forward(self, x):
        x = x.view(-1, 784) # Flatten the data (n, 1, 28, 28) -> (n, 784)
        x = F.relu(self.l1(x))
        x = F.relu(self.l2(x))
        x = F.relu(self.l3(x))
        x = F.relu(self.l4(x))
        return self.l5(x) # 마지막에는 activation function 통과시키지 않음. logits 으로 반환.
```

```python
model = Net()
model.to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.5)
```

```python
def train(epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % 100 == 0:
            print('Train Epoch: {} | Batch Status: {}/{} ({:.0f}% | Loss: {:.6f}'.format(
            epoch, batch_idx * len(data), len(train_loader.dataset),
            100. * batch_idx / len(train_loader), loss.item()))
```

```python
def test():
    model.eval()
    test_loss = 0
    correct = 0
    for data, target in test_loader:
        data, target = data.to(device), target.to(device)
        output = model(data)
        test_loss += criterion(output, target).item()
        #print(output.data.max(1, keepdim=True))
        pred = output.data.max(1, keepdim=True)[1]
        correct += target.data.view_as(pred).cpu().sum() # https://pytorch.org/docs/stable/generated/torch.eq.html : torch.eq
        print(correct)
    
    test_loss /= len(test_loader.dataset)
    print('===========================\n')
    print(f'Test set: Average loss: {test_loss:.4f}, Accuracy: {100*correct/len(test_loader.dataset)}')
```

```python
if __name__ == '__main__':
    since = time.time()
    for epoch in range(1, 2):
        epoch_start = time.time()
        train(epoch)
        m, s = divmod(time.time() - epoch_start, 60) # 몫과 나머지 구해줌. -> tuple 형태로 (몫, 나머지)
        print(f'Training time: {m:.0f}m {s:.0f}s')
        test()
        m, s = divmod(time.time() - epoch_start, 60)
        print(f'Testing time: {m:.0f}m {s:.0f}s')

    m, s = divmod(time.time() - since, 60)
    print(f'Total Time: {m:.0f}m {s:.0f}s\nModel was trained on {device}!')
```
