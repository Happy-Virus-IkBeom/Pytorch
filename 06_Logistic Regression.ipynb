```python
from torch import tensor
from torch import nn
from torch import sigmoid
import torch.nn.functional as F
import torch.optim as optim
```

```python
x_data = tensor([[1.0], [2.0], [3.0], [4.0]])
y_data = tensor([[0.], [0.], [1.], [1.]])
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear = nn.Linear(1,1)
        print(self.linear)
    
    def forward(self, x):
        y_pred = sigmoid(self.linear(x))
        return y_pred
```

```python
model = Model()
```

```python
criterion = nn.BCELoss(reduction='mean') #Binary Class Entropy
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(1000):
    y_pred = model.forward(x_data)
    loss = criterion(y_pred,y_data)
    print(f'Epoch {epoch +1}/1000 | Loss: {loss.data.item()}')
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
print(f'\nLet\'s predict the hours need to score above 50%\n{"="*50}')
hour_var = model.forward(tensor([[1.0]]))
print(f'Prediction after 1 hour of training: {hour_var.item():.4f} | Above 50%: {hour_var.item()  > 0.5}')
hour_var = model.forward(tensor([[7.0]]))
print(f'Prediction after 7 hour of training: {hour_var.item():.4f} | Above 50%: {hour_var.item()  > 0.5}')
```
