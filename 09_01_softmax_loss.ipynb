```python
from torch import nn, tensor, max
import numpy as np
```

```python
Y = np.array([1,0,0])
Y_pred1 = np.array([0.7, 0.2, 0.1])
Y_pred2 = np.array([0.1, 0.3, 0.6])
```

```python
print(f'Loss1: {np.sum(-Y*np.log(Y_pred1)):.4f}') # np.log 는 자연로그임, :.4f 는 소숫점 4자리까지 표현 cf) 밑이 10인 로그는 np.log10(x)
```

```python
print(f'Loss1: {np.sum(-Y*np.log(Y_pred2)):.4f}')
```

```python
# Input is class, not one-hot
# Y_pred 의 최댓값(Y_pred를 softmax 하고 확률의 최댓값)의 index 값의 정답. 
# 예를들어 [2,0,1]이면 차례대로 index 2번째 0번째 1번째 값이 정답으로 나와야함. Y_pred1 같은경우 정답인것이고 따라서 loss 낮음
# 반면 Y_pred2 는 logits 를 softmax 했을때 최댓값이 [0, 2, 2] 으로 나오니까 다 틀림. 따라서 loss가 높게 나온다.
Y = tensor([2, 0, 1], requires_grad=False) 
                                        
# input is of size nBatch x nClasses = 2 x 4
# Y_pred are logits (not softmax)
Y_pred1 = tensor([[0.1, 0.2, 0.9],
                  [1.1, 0.1, 0.2],
                  [0.2, 2.1, 0.1]])

Y_pred2 = tensor([[0.8, 0.2, 0.3],
                  [0.2, 0.3, 0.5],
                  [0.2, 0.2, 0.5]])

l1 = loss(Y_pred1, Y)
l2 = loss(Y_pred2, Y)
print(f'Batch Loss1:  {l1.item():.4f} \nBatch Loss2: {l2.data:.4f}')
```
