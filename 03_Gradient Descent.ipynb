```python
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
w = 1
l = 0.01
epoch = range(20)

w_list = []
loss_list = []
gradient_list = []
print(f"before training: 4hours -> {4*w} ")

def forward(x):
    return x * w

def loss(x,y):
    y_pred = forward(x)
    return (y_pred-y)**2

def gradient(x,y):
    return 2 * x * (x*w -y)

for _ in epoch:
    for x_val,y_val in zip(x_data,y_data):
        w = w - l * gradient(x_val, y_val)
    w_list.append(w)
    loss_list.append(loss(x_val,y_val))
    gradient_list.append(gradient(x_val,y_val))
print(f"after training: 4hours -> {4*w} ")

plt.plot(epoch, w_list, label='weight')
plt.plot(epoch, loss_list, label='loss')
plt.plot(epoch, gradient_list, label='gradient')
plt.legend()
plt.xlabel('epoch')
plt.show()
```
